```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
temperature_df = pd.read_csv('D:\Projetos\Python\Aprendendo Machine Learning\celsius-to-fahreinheit-machine-learning\Celsius-to-Fahrenheit.csv') #Atribuição do arquivo csv para a variável utilizando o pandas.
temperature_df.reset_index(drop = True, inplace = True) #Apagando indices caso existam no arquivo.
```

```python
x_train = temperature_df['Celsius']
y_train = temperature_df['Fahrenheit']
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, input_shape = [1]))
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss = 'mean_squared_error')
```

```python
epochs_hist= model.fit(x_train, y_train, epochs = 500)
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss']);
plt.title('Model loss progress during training')
plt.xlabel('Epoch')
plt.ylabel('Training loss')
plt.legend(['Training loss']);
```
